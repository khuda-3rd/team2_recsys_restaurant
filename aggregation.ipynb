### 가게의 위치 정보를 활용한 기준점(정문)으로부터의 거리 계산

```python
headers = {"Authorization": "KakaoAK c67ed56efd4280eb8e5f2e0be3ea737b"}
geocoding_url = 'https://dapi.kakao.com/v2/local/search/address.json?query='

def kakao_geocoder(address):
    try:
        url = geocoding_url + address
        result = json.loads(str(requests.get(url, headers=headers).text))
        match_first = result['documents'][0]['address']
        return pd.Series({'위도': float(match_first['y']), '경도': float(match_first['x'])})

    except:
        return None       
```

```python
jeongmoon=(37.247394,127.078381)
df['distance_정문']=0
```

```python
for idx in range(len(df)) :
    df['distance_정문'][idx]=haversine(jeongmoon,kakao_geocoder(df['addr'][idx]),unit='km')
```
